```python
taxi = df
taxi = taxi.rename(columns={'pcp 01': 'pcp_01', 'pcp 06': 'pcp_06', 'pcp 24': 'pcp_24' })
taxi.head()
```

```python
taxi['borough'].value_counts()
```

```python
taxi['pickups'].sum()
```

```python
taxi.groupby('borough').aggregate({'pickups': 'sum'})
```

```python
min_pickups = taxi.groupby('borough').pickups.sum().idxmin()
min_pickups
```

```python
min_pickups = taxi.groupby('borough').pickups.sum()
min_pickups
```

```python
taxi.groupby(['borough', 'hday']).agg({'pickups': 'mean'})
```

```python
pickups_by_mon_bor = (
    taxi.groupby(['pickup_month', 'borough'], as_index=False)
        .agg({'pickups' : 'sum'})
        .sort_values('pickups', ascending=False))
pickups_by_mon_bor
```
